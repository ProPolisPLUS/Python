```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"

page = requests.get(url)

soup = BeautifulSoup(page.text, "html")

#print(soup)
```

```python
table = soup.find_all("table")[1] # ima tri tabele na strani pa uzimamo drugu po redu pomocu indeksiranja 0,1,2
#print(table)
```

```python
column_title = table.find_all("th")
#print(column_title)
```

```python
"""
column_name = []
for title in column_title:
    title.text
    column_name.append(title)
print(column_name)

"""
#Laksi nacin
column_name = [title.text.strip() for title in column_title]
#print(column_name)
```

```python
df = pd.DataFrame(columns = column_name)
df
```

```python
trs = table.find_all("tr") #kompletni redovi u tabeli
#print(trs)
```

```python
for i in trs:
    rows = i.find_all("td")
    individual_row = [data.text.strip() for data in rows]
    if len(individual_row) > 0:
        df.loc[len(df)] = individual_row
        print(individual_row)

'''Ovdje i predstavlja jedan red u tabeli,
 zatim se stvara promenljiva sa svim td stavkama,
zatim svaka da td stavka se transformise u text i stripuje unutrar liste,
 a zatim se lista printa ako je ispunjen uslov da lista nije prazna'''
```
