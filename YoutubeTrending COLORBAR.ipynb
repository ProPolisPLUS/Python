```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("fivethirtyeight")

df = pd.read_csv("C:/Users/Nikola/Desktop/Nikola-Programi/CSV-za-vjezbu/CC-python-pandas-matplotlib-master/2019-05-31-data.csv")

view_count = df["view_count"]
likes = df["likes"]
ratio = df["ratio"]

plt.scatter(view_count,likes, ec="Black", alpha = 1, lw=1, c=ratio, cmap="summer")
# lw je debljina edgecolor - a, tj. obima krugova
# cmap boja color mape koja se nalazi sa strane grafika

cbar = plt.colorbar()
# dodavanje color bara sa strane grafika
cbar.set_label("Like/Dislike Ratio")
# dodavanje naziva colorbara

plt.xscale("log")
plt.yscale("log")
'''Funkcija se koristi za promjenu skale i razmjer na osi, medjutim'''
plt.title("Youtube Tranding Videos")
plt.xlabel("Views")
plt.ylabel("Likes")

plt.ylim(10**3,10**(7.2))


plt.tight_layout()






plt.show()
```
