```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5,6]
y = [3,6,2,4,3,6,]

plt.style.use("seaborn-v0_8-paper")
#izgled grafika
#za brze biranje mozemo ukucati plt.style.use(style), ova funckija ce svakim pokretanjem programa mijenjati izgled
# kako bi znali kako se zove grafik koji nam se dopadne  dodamo funkciju plt.figure(style), tad ce ime grafika biti i izabrani izgled

plt.plot(x,y, color="m", marker="d", mfc="Black", mec="m", ms=10, ls="-")
''' marker - dodaje se na svaku tacku neki  oblik, "d" je diamond, za vise oblika pogledaj na internet
    mef-marker face color
    mec- marker edge color
    ms- marker size
    ls- line style, linestyle takodje mozes vidjeti na internet, matplot lib sajtu'''

plt.xlim(0,7)
plt.ylim(0,7)

d={"backgroundcolor": "lightblue", "color":"black", "family":"Times New Roman", "size":9, "style":"italic", "weight":"bold" }
# d={}, dictionary, naredbe kako ce izgledati dodati zapis

plt.text(0.2,3.5, "Prva tacka",fontdict = d)
# plt.text() prvo se dodaju kordinate teksta, zatim zapis pod znacima navoda, i poziva se naredba odnosno opis izgleda texta
plt.title("Grafik za vjezbanje")
plt.xlabel("X osa")
plt.ylabel("Y osa")



plt.hlines(y=3,xmin=0, xmax=1, color="grey", ls="--" )
plt.vlines(x=1,ymin=0, ymax=3, color="grey", ls="--")
#Dodavanje horizontalnih i vertikalnih linija u granicama

plt.axhline(y=1.5, color="red", ls="--")
plt.axvline(x=5.5, color="Blue", ls="--")
# dodavanje horizontalnih i vertikalnih linija

plt.xticks([1.5,4,7],["A","B","C"])
# podesavanje x ose, umjesto brojeva 1.5,4,7 stavi slova A,B,C
plt.yticks(np.arange(0,7,0.5))
# podesavanje y ose, podesi osu od 0 do 7 tako da su razmaci po 0.5

print(plt.style.available)
# ova linija nam isprinta moguce izglede grafika
plt.show()
```
