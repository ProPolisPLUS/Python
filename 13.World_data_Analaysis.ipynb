```python
#Jos jedan primjer

data = {'Category': ['A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
        'Value': [10, 15, 20, 25, 30, 35, 40, 45, 50]}

df = pd.DataFrame(data)

# Postavljanje veličine grafa
plt.figure(figsize=(8, 6))

# Crtanje boxplot grafa sa Seaborn
sns.boxplot(x='Category', y='Value', data=df, palette='Set1')

# Dodavanje oznaka osi i naslova
plt.xlabel('Kategorija')
plt.ylabel('Vrijednost')
plt.title('Boxplot po kategoriji')

# Prikaz grafa
plt.show()
```
