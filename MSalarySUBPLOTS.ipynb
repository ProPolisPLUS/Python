```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("fivethirtyeight")

df = pd.read_csv("C:/Users/Nikola/Desktop/Nikola-Programi/CSV-za-vjezbu/CC-python-pandas-matplotlib-master/data.csv")

ages = df["Age"]
devs = df["All_Devs"]
py = df["Python"]
js = df["JavaScript"]

fig, (ax1,ax2) = plt.subplots(nrows = 2, ncols = 1)

#fig1, ax1 = plt.subplots() '''Ukoliko nam trebaju 2 grafika zasebno u dvije slike'''
#fig2, ax2 = plt.subplots()

'''Sto se god odnosi na posebne grafike koristi se ax1/2...
rwos i cols se dodaju da bi podesio oce li grafici bit jedno iznad drugog ili jedno pored drugog'''


ax1.plot(ages, devs, label="All Devs", ls="--", color= "Black")
ax2.plot(ages, py, label = "Python")
ax2.plot(ages, js, label = "JavaScript")



ax1.set_yticks(range(10000,120000,20000))
ax2.set_yticks(range(10000,120000,20000))


ax1.set_title("Median salary")
ax1.set_xlabel("Ages")
ax1.set_ylabel("Salary")
ax1.legend()


ax2.set_xlabel("Ages")
ax2.set_ylabel("Salary")
ax2.legend()

plt.tight_layout()
plt.show()


```
