```python
import matplotlib.pyplot as plt
x = ["Huawai","Samsung","Apple","LG"]
y = [5,2,7,1]
z = [0.1,0,0,0]

plt.pie(y, labels=x, explode=z, autopct=lambda y:f"{round(y,2)} %")
''' autopct = lambda y: f"{Promjenljive}%"
znaci, definisi funkciju lambda koja ce y da promijeni u procenat.
round(y,2) znaci da promjenljivu y zaokruzi na dvije decimale. 
autopct je ime lambda funkcije, lambda je trenutna funkcija koju cemo koristiti
y:y, lijevo y je y od unosa a desno od output-a, pa tako podesavamo ono sto ce izaci...U nasem slucaju izaci ce procenat zaokruzen na dvije decimale '''
plt.show()
```
