```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")

df = pd.read_csv("C:/Users/Korisnik/Desktop/DataAnalyst/Python/Podaci-za-analizu/CSV-za-vjezbu/CC-python-pandas-matplotlib-master/data.csv")

#godine i prosjecne zarade
ages = df["Age"]
python = df["Python"]
Java = df["JavaScript"]
all = df["All_Devs"]

plt.plot(ages, python, label="Python")
plt.plot(ages,all, color="Black", ls="--", label = "All Devs")

plt.fill_between(ages,python,all, where=(python > all), interpolate=True, alpha=0.5, label="Above AVG")
plt.fill_between(ages,python,all, where=(python <= all), interpolate=True, alpha=0.5, label="Below AVG")
'''-Dodajemo na prva dva mjesta izmedju cega ce da bude prostor ispunjen bojom ages i python
-na trece mjesto definisemo granicu, da bude obojano iznad i ispod all promjenjive
-zatim dodajemo tacna podrucija, ako je iznad oboj plavom a ako je ispod crvenom
-onda dodajemo interpolaciju, ona omogucava da na grafik nema neobojenih rupa
-Alpa je nijansa boje
!python prvi grafik sam boji plavom a drugi crvenom pa je nepotrebno dodavati color=""'''

plt.title("Median salary (USD) by age")
plt.xlabel("Ages")
plt.ylabel("Salary")
plt.tight_layout()
plt.legend()
plt.show()



```
