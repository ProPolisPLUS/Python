```python
#Ajde sad da uzmemo malo bolji prijmjer, napravio sam country_grp dje je grupa svih zemalja na anketi
country_grp = df.groupby(['Country'])
#Sad cemo da vidimo prosjecnu platu u Njemackoj
country_grp['ConvertedCompYearly'].median().loc['Germany']
```

```python
country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
#Koristili smo apply metodu, kao i lambda funkciju od koje smo trazili da nam povratna informacija bude broj ljudi koji znaju Python

```

```python
#Zanima nas koji procenar ljudi zna pajton iz koje zeblje, to cemo uraditi ovako.
country_uses_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
#Napravicemo promjenljivu koja predstavlja ljude koji koriste pyhon, i promjenljivu koja nam kaze koliko je ljudi iz koje drzavee
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
#Zatim cemo spojiti ove dvije kolone u jednu tabelu, ovom funkcijom
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
#Mora se dodati axis jer pajton automacki spaja redavoe a ne kolone
python_df
```

```python
#Oromijenicemo imena kolona da je jasnije
python_df.rename(columns={'count': 'NumRespondents', 'LanguageHaveWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
#Sada pravimo novu kolonu i racunamo procenat
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
python_df
```

```python
#Zatim mozemo sortirati zemlje od najveceg do najmanjeg procenta
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
python_df
```
