```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress




#Ucitavamo fajl sa kompjutera sledecom komandom
'''Da bi mogao da se ucita fajl, potrebno je fajl snimiti u istom direktorijumu gdje je i jupyter
da bi provjerili dje je jupyter kucamo funkciju
import os
print(os.getcwd)'''
'''Moze se unijet i putanja dajla ali mora se navesti tacna putanja sa ovakvim crticama / '''
df = pd.read_csv("C:/Users/Nikola/Desktop/Nikola-Programi/CSV-za-vjezbu/CC-python-pandas-matplotlib-master/scottish_hills.csv")
df1 = df.head(30) #uzeo sam samo prvih 30 tacaka
x = df1.Height   # sto hocu da mi bude x a sto y osa
y = df1.Latitude

plt.style.use("seaborn-v0_8-darkgrid")


plt.scatter(x,y, s=10, label="Tacke na grafiku", marker = "x", color = "m")
plt.title("Vjezbanje")
plt.xlabel("Visina planina")
plt.ylabel("Geografska sirina")

stats = linregress(x, y) #formiram linearnu regresiju

m = stats.slope #uzmem da je nagim linearne regresije m
b = stats.intercept # uzmem da je presjek linearne regresije b
r2=x.corr(m*x+b) # sintaksa za linearnu regresiju

plt.plot(x,m*x+b, color="Black", label = "Linearna regresija", lw=1) # formiram pravu liniju od promjenjljivih
d={"backgroundcolor": "lightblue", "color":"black", "family":"Times New Roman", "size":9, "style":"italic", "weight":"bold" }
plt.text(1150,57.6,"Faktor korelacije:",fontdict = d)
plt.text(1220,57.6,r2,fontdict = d)


plt.legend()
plt.show()





print("Nagib linearne regresije je :", m)
print("Presjek linearne regresije je :", b)
#print(plt.style.available)

```
